```python
#rename column
df2["totalReviewers"] = df2["reviewerID_y"]
df2["overallScore"] = df2["overall_x"]
df2["summaryReview"] = df2["summary_x"]

df2_smp = df2[['asin','summaryReview','overallScore',"totalReviewers"]]
df2_smp.head()
```

```python
df3 = df.groupby('asin').mean()
df3.head()
```

```python
ProductReviewSummary = df2.groupby("asin")["summaryReview"].apply(list)
ProductReviewSummary = pd.DataFrame(ProductReviewSummary)
ProductReviewSummary = ProductReviewSummary.reset_index()
ProductReviewSummary.head()
```

```python
df4 = pd.merge(ProductReviewSummary, df3, on=['asin'], how = 'inner')
df4.head()
```

```python
df4['summaryReview'] = [','.join(map(str, l)) for l in df4['summaryReview']]
```

```python
df4 = df4[['asin', 'summaryReview', 'overall', 'totalReviewers']]
```

```python
#function for tokenizing summary

import re
regEx = re.compile('[^a-z]+')
def cleanReviews(reviewText):
    reviewText = reviewText.lower()
    reviewText = regEx.sub(' ', reviewText).strip()
    return reviewText
```

```python
import string

df4["summaryReview"] = df4["summaryReview"].apply(cleanReviews)
df4 = df4.drop_duplicates(['overall'], keep='last')
df4 = df4.drop(['index'], axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
df4['summaryReview'] = df4['summaryReview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df4['summaryReview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df4.index, index=df4['asin']).drop_duplicates()
```

```python
df4.to_csv('df4.csv', index = False)
```

```python
 #Function that takes in product id as input and gives recommendations 
def content_recommender(asin, cosine_sim=cosine_sim, df=df4, indices=indices):
    # Obtain the index of the product that matches the asin
    idx = indices[asin]

    # Get the pairwsie similarity scores of all products with that product
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the products based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar products. Ignore the first movie.
    sim_scores = sim_scores[1:11]

    # Get the product indices
    product_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar products
    return df['asin'].iloc[product_indices]
```
